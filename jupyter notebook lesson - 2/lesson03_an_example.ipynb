## Ví dụ minh họa phương pháp biến đổi ngược trên lớp 

Chọn ngẫu nhiên một điểm trong hình vuông có tâm tại gốc tọa độ và chiều dài cạnh là 2, gọi $R$ là khoảng cách từ điểm được chọn đến tâm hình vuông. Viết hàm sinh $R$.

Nếu được dùng 2 số ngẫu nhiên có phân phối đều trong đoạn [0, 1] (tức là được phép gọi 2 lần hàm `np.random.rand`) thì dễ.

```python
def R():  
    X = 2*np.random.rand()
    Y = 2*np.random.rand()
    return np.sqrt((1 - X)**2 + (1 - Y)**2)
```

Nếu chỉ được dùng 1 số ngẫu nhiên có phân phối đều trong đoạn [0, 1] thì khó. Dùng phương pháp biến đổi ngược. Tính hàm phân phối tích lũy của $R$, là $F$ sau đó tìm hàm ngược $F^{-1}$ và sinh $R$.

```python
r = np.linspace(1, np.sqrt(2), 100)
u = np.pi*r**2/4 - r**2 * np.arccos(1/r) + np.sqrt(r**2 - 1)
plt.plot(r, u)
plt.show()
```

```python
def inv(u, eps=10**-6):
    l, r = 1, np.sqrt(2)
    while np.abs(r - l) > eps:
        m = (l + r)/2
        v =  np.pi*m**2/4 - m**2 * np.arccos(1/m) + np.sqrt(m**2 - 1)
        if v > u:
            r = m
        else:
            l = m
    return (l + r)/2
```

```python
inv(0.9)
```

```python
def R2():
    U = np.random.rand()
    if 0 <= U <= np.pi/4:
        return 2*np.sqrt(U/np.pi)
    return inv(U)
```

Dùng histogram để kiểm tra kết quả: `R` và `R2` sinh ra các giá trị cùng phân phối như yêu cầu.

```python
plt.hist([R() for _ in range(100000)], bins=50)
plt.show()
```
